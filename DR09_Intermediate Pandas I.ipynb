```python
import pandas as pd
```

```python
penguins = pd.read_csv("penguins.csv")
```

```python
# information dataframe
penguins.info()
```

```python
# select column
penguins[['island','sex']]
penguins.island
```

```python
# iloc
penguins.iloc[ 0:2, [1,2,3] ]
```

```python
# filter df with one condition
penguins[penguins['island'] == 'Torgersen']
```

```python
penguins[penguins['bill_length_mm'] < 34]
```

```python
# filter df with multi conditions
penguins[(penguins['bill_length_mm'] < 34) & (penguins['island'] == 'Torgersen')]
```

```python
# fitler with query
penguins.query('(island == "Torgersen") | (island == "Dream")')
penguins.query('island == "Torgersen"')
```

```python
# check missing values
penguins.isna().sum()
```

```python
# filter missing values
penguins[penguins['sex'].isna()]
```

```python
# fill missing values
# Mean Imputation
mean_bill = penguins['bill_length_mm'].mean()
```

```python
# fill value
penguins['bill_length_mm'].fillna(mean_bill)
```

```python
# sorting
penguins.sort_values('bill_length_mm', ascending=False)
```

```python
# unique and counting
penguins['species'].unique()
penguins['species'].value_counts()
```
