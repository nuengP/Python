```python
print(f"Total : \n{Nation[['Number of Arrivals - Chiangrai','Number of Arrivals - Thailand']].sum()}")

# Calculate the proportion of arrivals in Chiang Rai compared to the total arrivals in Thailand
perc_CEI = round(Nation['Number of Arrivals - Chiangrai'].sum()/ Nation['Number of Arrivals - Thailand'].sum(), 3) *100
print(f"Percent of Tourism arrvials to Chaing Rai by total is : {perc_CEI} %")
```

```python

# Create data for pie chart
labels = ['Arrivals in Chiang Rai', 'Remaining Arrivals in Thailand']
sizes = [perc_CEI, 100 - perc_CEI]

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['#CC66FF', '#FF0000'], textprops={'fontsize': 15, 'color': 'black'})
plt.title('Percentage of Tourism Arrivals to Chiang Rai Compared to Total Arrivals in Thailand')
plt.axis('equal')
```

```python
## Exploring why Chiang Rai isn't attracting as many tourists as other parts of Thailand.
Prop = Nation['Number of Arrivals - Chiangrai']/ Nation['Number of Arrivals - Thailand']
Prop_df = pd.DataFrame({'Country' : Nation['Country of Residence'], 'Proportion' : Prop})
Prop_df = Prop_df[Prop_df['Proportion'] > 0]
Top_10_Prop = Prop_df.sort_values(by = 'Proportion').head(10)
Top_10_Prop
```

```python
Top_10_Prop['Percent'] = Top_10_Prop['Proportion']*100

# bar chart 1 -> top bars (overall tourism in Thailand)
bar1 = sns.barplot(y="Country",  x="Percent", data=Top_10_Prop, color='darkblue')

```

### To boost tourist arrivals in Chiang Rai, we can target these 5 countries: Cambodia, India, Vietnam, Russia, and Indonesia.

```python
# Calculate the total number of arrivals in Chiangrai by region
arrivals_by_region = Nation[Nation['Number of Arrivals - Chiangrai'] > 1000].groupby('Region')['Number of Arrivals - Chiangrai'].sum()

arrivals_by_region
```

```python


# Create pie chart
colors = sns.color_palette("deep")
plt.figure(figsize=(8, 8))
plt.pie(arrivals_by_region, labels=arrivals_by_region.index, autopct='%1.1f%%', startangle=140, labeldistance=0.9, textprops={'fontsize': 12, 'color': 'black'})
plt.title('Distribution of Arrivals in Chiangrai by Region')

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')
```

### Europe and East Asia have been the most popular regions for arrivals in Chiang Rai.
